```python
from sklearn.model_selection import train_test_split

# Select only Age, Sex, ChestPain, RestBP, Chol
columns = ['Age', 'Sex', 'ChestPain', 'RestBP', 'Chol']
data_subset = df[columns]

# Split into training (75%) and testing (25%)
train_data, test_data = train_test_split(data_subset, test_size=0.25, random_state=42)

print("Training data shape:", train_data.shape)
print("Testing data shape:", test_data.shape)

# Show first few rows of training data
train_data.head()
```

```python
# Confusion Matrix values
TP = 45
FP = 100 - 45
FN = 50 - 45
TN = 500 - (TP + FP + FN)

# Display Confusion Matrix
import pandas as pd
conf_matrix = pd.DataFrame([[TP, FP],
                            [FN, TN]],
                           columns=['Predicted Positive', 'Predicted Negative'],
                           index=['Actual Positive', 'Actual Negative'])
print("Confusion Matrix:\n", conf_matrix)

# Compute metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)

print("\nEvaluation Metrics:")
print(f"Accuracy:  {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall:    {recall:.2f}")
print(f"F1 Score:  {f1_score:.2f}")
```
